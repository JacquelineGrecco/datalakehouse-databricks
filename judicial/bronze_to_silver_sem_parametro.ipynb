```python
from pyspark.sql.functions import lit, current_date

catalog_source = 'dev_bronze'
catalog_target = 'dev_silver'

table_path = f'{catalog_source}.sql_server.customers'
table_target = f'{catalog_target}.judicial.customers'

print(table_path)
print(table_target)

df = spark.read.table(table_path)

df = df.withColumnRenamed('email', 'email_address')
df = df.withColumn('dt_load', lit(current_date()))

df.write.mode('overwrite').saveAsTable(table_target)
```
